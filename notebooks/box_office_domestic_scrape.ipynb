```python
#2017
us_17 = pd.read_html('https://www.boxofficemojo.com/daily/2017/?view=year')

us_17 = us_17[0]
us_17_df = pd.DataFrame(us_17)

us_17_df['Year'] = '2017'
us_17_df['Country'] = 'America'

#2018
us_18 = pd.read_html('https://www.boxofficemojo.com/daily/2018/?view=year')

us_18 = us_18[0]
us_18_df = pd.DataFrame(us_18)

us_18_df['Year'] = '2018'
us_18_df['Country'] = 'America'

#2019
us_19 = pd.read_html('https://www.boxofficemojo.com/daily/2019/?view=year')

us_19 = us_19[0]
us_19_df = pd.DataFrame(us_19)

us_19_df['Year'] = '2019'
us_19_df['Country'] = 'America'

#2020
us_20 = pd.read_html('https://www.boxofficemojo.com/daily/2020/?view=year')

us_20 = us_20[0]
us_20_df = pd.DataFrame(us_20)

us_20_df['Year'] = '2020'
us_20_df['Country'] = 'America'

#2021
us_21 = pd.read_html('https://www.boxofficemojo.com/daily/2021/?view=year')

us_21 = us_21[0]
us_21_df = pd.DataFrame(us_21)

us_21_df['Year'] = '2021'
us_21_df['Country'] = 'America'

#2022
us_22 = pd.read_html('https://www.boxofficemojo.com/daily/2022/?view=year')

us_22 = us_22[0]
us_22_df = pd.DataFrame(us_22)

us_22_df['Year'] = '2022'
us_22_df['Country'] = 'America'

#concatenated data frame
uss = [us_17_df, us_18_df, us_19_df, us_20_df, us_21_df, us_22_df]
  
box_office_domestic = pd.concat(uss)
box_office_domestic
```
